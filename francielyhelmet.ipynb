```python
data_yaml = """
path: /content/drive/MyDrive/Helmet_Detection1.v2i.yolov8

train: train/images
val: valid/images
test: test/images

names:
  0: Full Face
  1: Helmet
  2: No Helmet
"""

with open("data.yaml", "w") as f:
    f.write(data_yaml)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data="data.yaml",
    epochs=20,
    imgsz=384,
    batch=16,
    project="/content/drive/MyDrive/Helmet_Detection1.v2i.yolov8",
    name="yolov8_epi_model"
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/Helmet_Detection1.v2i.yolov8/yolov8_epi_model4/weights/best.pt")

metrics = model.val(
    data="/content/drive/MyDrive/Helmet_Detection1.v2i.yolov8/data.yaml",
    split='test',
    plots=True
)

print("Precisão por classe:", metrics.box.p)
print("Recall por classe:",    metrics.box.r)
print("F1-score por classe:",  metrics.box.f1)

print("mAP@0.5 por classe:", metrics.box.ap50)
print("mAP@0.5:0.95 por classe:", metrics.box.ap)
print("mAP@0.5 geral:", metrics.box.map50)
print("mAP@0.5:0.95 geral:", metrics.box.map)
```
